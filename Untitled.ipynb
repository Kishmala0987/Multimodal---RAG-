## Separate Tables from Text

```python
tables = []
texts = []
for i,chunk in enumerate(chunks):
    texts.append(chunk)
    if chunk.metadata.orig_elements:
        for el in chunk.metadata.orig_elements:
            if "Table" in str(type(el)):
                tables.append(el)
    
```

## Get Images

```python
def get_images_base64(chunks):
    image_base64 = []
    for chunk in chunks:
        for el in chunk.metadata.orig_elements:
            if "Image" in str(type(el)):
                image_base64.append(el.metadata.image_base64)

    return image_base64

images = get_images_base64(chunks)
```

```python
class MultiModalRetriever:
    def __init__(self, vectorstore, docstore, id_key="doc_id"):
        self.vectorstore = vectorstore.as_retriever()
        self.docstore = docstore
        self.id_key = id_key

    def get_relevant_documents(self, query):
        hits = self.vectorstore.invoke(query)
        key = [hit.metadata[self.id_key] for hit in hits]
        results = self.docstore.mget(key)
        return results
```

```python
retriever = MultiModalRetriever(
    vectorstore=vectorstore,
    docstore=docstore,
    id_key=id_key,
)
```

```python
query = "What is attention mechainsim?"
results = retriever.get_relevant_documents(query)
#results = retriever.invoke(query)
for r in results:
    print(r)
```

# RAG pipeline

```python
from langchain_core.runnables import RunnablePassthrough, RunnableLambda
from langchain_core.messages import SystemMessage, HumanMessage
from base64 import b64decode
```

```python
#detect if image otherwise add it in text on retrieved documents
import base64

def parse_docs(docs):
    b64 = []
    text = []
    for doc in docs:
        try:
            base64.b64decode(doc, validate=True)
            b64.append(doc)
        except Exception:
            text.append(doc)
    return {"image": b64, "text": text}
    
```

```python


def build_prompt(kwargs):
    """
    kwargs expected:
    {
        "context": {"text": [...], "image": [...]},
        "question": "user question string"
    }
    Returns a ChatPromptTemplate ready to send to the LLM.
    """

    docs_by_type = kwargs["context"]

    # 1️⃣ Concatenate all text elements into context_text
    context_text = ""
    for t in docs_by_type.get("text", []):
        if hasattr(t, "text"):
            context_text += t.text + "\n"
        elif isinstance(t, str):
            context_text += t + "\n"

    # 2️⃣ Build the base prompt template (keep placeholders, no f-string)
    prompt_template = """
Answer the question based only on the following context, which may include text, tables, and image summaries.
If the answer cannot be found in the context, say "I don't know."

Context:
{context_text}

Question:
{user_question}
"""

    # 3️⃣ Add images if any (still with placeholders)
    if docs_by_type.get("image"):
        prompt_template += "\n\nThe following images are included (base64 encoded):\n"
        for i, img_b64 in enumerate(docs_by_type["image"], start=1):
            prompt_template += f"Image {i+1}: data:image/jpeg;base64,{img_b64}\n"

    # 4️⃣ Create ChatPromptTemplate from the string (no input_variables)
    chat_prompt = ChatPromptTemplate.from_template(prompt_template)

    # 5️⃣ Return **formatted prompt** ready for the LLM
    # Note: Here we format it immediately
    return chat_prompt.format(context_text=context_text, user_question=kwargs["question"])
```

```python
model = ChatOpenAI(
    model = "gpt-4o-mini"
)
parser = StrOutputParser()
chain = (
    {
        "context": RunnableLambda(lambda x: retriever.get_relevant_documents(x)) | RunnableLambda(parse_docs),
        "question": RunnablePassthrough()
    } | RunnableLambda(build_prompt) | model | parser
)
```

```python
response = chain.invoke("What is attention mechanism?")
print(response)
```

```python
chain_with_sources = (
    {
        "context": RunnableLambda(lambda x: retriever.get_relevant_documents(x)) | RunnableLambda(parse_docs),
        "question": RunnablePassthrough()
    } | RunnablePassthrough().assign(
        response=(
            RunnableLambda(build_prompt) | model | parser
        )
    )
)
```
